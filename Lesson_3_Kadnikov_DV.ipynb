### Решение задание №1

$$1.\,\,\,\{a_n\}_{n=1}^\infty=2^n-n$$
<br>
a. Исследовать их на монотонность:
<br>
$a_1 = 2^1 - 1 = 1$
<br>
$a_2 = 2^2 - 2 = 2$
<br>
$a_3 = 2^3 - 3 = 5$
<br>
<br>
$a_{1}\lt a_{2}\lt a_{3}$ - последовательность монотонно возрастает
<br>

b. Исследовать на ограниченность:
<br>
Оценка ограничения снизу при $n=1: \,\,\, a_1 = 2^1-1 = 1$
<br>
Оценка ограничения снизу при $n\to \infty$:
$$\lim_{n\to \infty} (2^n-n) = \lim_{n\to \infty} \Bigr((2^n-n)\cdot \frac{2^n+n}{2^n+n}\Bigl) = \lim_{n\to \infty} \Bigr( \frac{2^{2n}-n^2}{2^n+n} \Bigl) = \lim_{n\to \infty} \frac{2^n\Bigr(2^{2}-\frac{n^2}{2^n}\Bigl)}{2^n\Bigr(1-\frac{n}{2^n}\Bigl)} = \frac{4}{1} = 4$$
<br>
Последовательность ограничена снизу в значении $1$, сверху в значении $4$
<br>
<br>
c. Найти пятый по счету член:
<br>
$$a_5 = 2^5 - 5 = 27$$
<br>

$$2.\,\,\,\{b_n\}_{n=2}^\infty=\frac{1}{1 - n}$$
<br>
a. Исследовать их на монотонность:
<br>
$b_2 = \frac{1}{1 - 2} = -1$
<br>
$b_3 = \frac{1}{1 - 3} = -\frac{1}{2}$
<br>
$b_4 = \frac{1}{1 - 4} = -\frac{1}{3}$
<br>
<br>
$b_2\lt; b_3\lt; b_4$ - последовательность монотонно возрастает
<br>
<br>
b. Исследовать на ограниченность:<br>
<br>
Оценка ограничения снизу при $n=2: \,\,\, b_2 = \frac{1}{1 - 2} = -1$<br>
<br>
Оценка ограничения снизу при $n\to \infty$:
<br>
$$\lim_{n\to \infty} \frac{1}{1 - 2} = 0$$ Последовательность ограничена снизу в значении $-1$, сверху в значении $0$
<br>
c. Найти пятый по счету член:
<br>
$$b_5 = \frac{1}{1 - 5} = -\frac{1}{4}$$
<br>

$$3.\,\,\,\{c_n\}_{n=1}^\infty=(-1)^n + \sqrt{2n}$$
<br>
a. Исследовать их на монотонность:<br>
<br>
$c_1 = (-1)^{1} + \sqrt{2\cdot 1} = -1 + \sqrt{2} \approx 0,4$<br>
$c_2 = (-1)^2 + \sqrt{2\cdot 2} = 1 + 2 = 3$<br>
$c_3 = (-1)^3 + \sqrt{2\cdot 3} = -1 + \sqrt{6} \approx 1,45$<br>

$c_1\lt; c_2\gt; c_3$ - последовательность немонотонная<br>
<br>
b. Исследовать на ограниченность:
<br>
Оценка ограничения снизу при $n=1: \,\,\, c_1 = (-1)^{1} + \sqrt{2\cdot 1} = -1 + \sqrt{2}$
<br>
Оценка ограничения снизу при $n\to \infty$:<br>
<br>
$$\lim_{n\to \infty} ((-1)^n + \sqrt{2n}) = \lim_{n\to \infty} \frac{((-1)^n + \sqrt{2n})\cdot ((-1)^n - \sqrt{2n})}{(-1)^n - \sqrt{2n}} = \frac{(-1)^{2n} - 2n}{(-1)^n - \sqrt{2n}} = \infty$$
<br>

Последовательность неограничена<br>
<br>
c. Найти пятый по счету член:
<br>
$$c_5 = (-1)^5 + \sqrt{2\cdot 5} = -1+ \sqrt{10} \approx 2,16$$
<br>

$$4.\,\,\,\{d_n\}_{n=1}^\infty=(-1)^{2n} + \frac{1}{n^2}$$
<br>
a. Исследовать их на монотонность:
<br>
$d_1 = (-1)^{2} + \frac{1}{1^2} = 1 + 1 = 2$<br>
$d_2 = (-1)^{4} + \frac{1}{2^2} = 1 + 0,25 = 1,25$<br>
$d_3 = (-1)^{6} + \frac{1}{3^2} = 1 + \frac{1}{9} = 1\frac{1}{9}$<br>

$d_1\gt; d_2\gt; d_3$ - последовательность монотонно убывает<br>
<br>
b. Исследовать на ограниченность:<br>
<br>
Оценка ограничения снизу при $n=1: \,\,\, d_1 = (-1)^{2} + \frac{1}{1^2} = 1 + 1 = 2$<br>
<br>
Оценка ограничения снизу при $n\to \infty$:<br>
<br>
$$\lim_{n\to \infty} \bigr((-1)^{2n} + \frac{1}{n^2}\bigl) = 1 + 0 = 1$$ Последовательность ограничена снизу в значении $2$, сверху в значении $1$<br>
<br>
c. Найти пятый по счету член:<br><br>
<br>
$$d_5 = (-1)^{10} + \frac{1}{5^2} = 1 + \frac{1}{25} = 1,04$$
<br>
